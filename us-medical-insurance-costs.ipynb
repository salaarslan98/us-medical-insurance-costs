```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def average(number_list):
    sum = 0
    for num in number_list:
        sum += num
    
    average_num = sum/len(number_list)
    
    return average_num
```

```python
age_list = []
charge_list = []
bmi_list = []
sex_list = []
smoker_list = []
children_list = []
region_list = []
with open('insurance.csv', newline = '\n') as insurance_list:
    insurance_list_data = csv.DictReader(insurance_list)
    for row in insurance_list_data:
        age_list.append(int(row['age']))
        charge_list.append(float(row['charges']))
        bmi_list.append(float(row['bmi']))
        sex_list.append(row['sex'])
        smoker_list.append(row['smoker'])
        children_list.append(int(row['children']))
        region_list.append(row['region'])
        

print("The average age of all the users in the data set is {}".format(round(average(age_list))))
print("The average charges incurred by all the users in the data set is {} USD".format(round(average(charge_list),2)))
# print(bmi_list)
```

```python
def dictionary_of_combined_data(age_list,children_list,bmi_list, smoker_list, region_list, sex_list, charge_list):
    combined_data = {}
    combined_data_list = []
    for num in range(len(smoker_list)):
        data = dict()
        data["Age"] = age_list[num]
        data["Number of Children"] = children_list[num]
        data["BMI"] = bmi_list[num]
        data["Smoker"] = smoker_list[num]
        data["Region"] = region_list[num]
        data["Sex"] = sex_list[num]
        data["Charges"] = charge_list[num]
        combined_data_list.append(data)
    
    return combined_data_list

insurance_list = dictionary_of_combined_data(age_list,children_list,bmi_list, smoker_list, region_list, sex_list, charge_list)
insurance_list
```

```python
def key_dict_values(combined_data, list): 
    dictionary_list = []
    for data in combined_data:
        for dictionary in data.values():
            if dictionary == list:
                dictionary_list.append(data)
    
    return dictionary_list
```

```python
dictionary_region = {}
for data in region_list:
    dictionary_region[data] = key_dict_values(insurance_list, data)

dictionary_region
```

```python
dictionary_smoker = {}
for data in smoker_list:
    dictionary_smoker[data] = key_dict_values(insurance_list,data)

dictionary_smoker
```

```python
dictionary_children = {}
for data in children_list:
    dictionary_children[data] = key_dict_values(insurance_list,data)

dictionary_children
```

Smokers based on gender (done)

```python
sex_smoker_list = list(zip(sex_list, smoker_list))
female_smoker = 0
male_smoker = 0
male_nonsmoker = 0
female_nonsmoker = 0
for item in sex_smoker_list:
    if item == ('female', 'yes'):
        female_smoker += 1
    elif item == ('female', 'no'):
        female_nonsmoker += 1
    elif item == ('male', 'yes'):
        male_smoker += 1
    else: 
        male_nonsmoker += 1
# print(sex_smoker_list)

print("There are {} female smokers and {} male smokers in this data set".format(female_smoker, male_smoker))
print("There are {} female and {} male nonsmokers in this data set".format(female_nonsmoker, male_nonsmoker))

sex_smoker_dict = {'Male Smoker': male_smoker/len(smoker_list)*100, 'Male Nonsmoker': male_nonsmoker/len(smoker_list)*100, 'Female Smoker': female_smoker/len(smoker_list)*100, 'Female Nonsmoker': female_nonsmoker/len(smoker_list)*100}
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(sex_smoker_dict.keys(),sex_smoker_dict.values())
plt.show()
```

```python
plt.pie(sex_smoker_dict.values(), labels = sex_smoker_dict.keys())
plt.show()
print("From this we can see the true representation of female and male smokers to non smokers in our dataset")
```
